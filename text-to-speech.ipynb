# microsoft/speecht5_tts - быстрый, но не берет русский

```python
# Загружаем xvector, который отвечает за характеристики голоса спискера.
from datasets import load_dataset
embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")

speaker_embedding = torch.tensor(embeddings_dataset[7306]["xvector"]).unsqueeze(0)
```

```python
# Задаем пайплайн - вид модели, который отвечает за выполнение задачи.
# В нашем случае это будет "текст в речь".
synthesizer = pipeline("text-to-speech", "microsoft/speecht5_tts")
```

```python
# Передаем фразу и характеристики голоса синтезатору. Затем создаем звуковой файл.
# Файл появится на панели слева - жмякайте на символ папки.
speech = synthesizer("Hello, my dog is cooler than you!", forward_params={"speaker_embeddings": speaker_embedding})

sf.write("speech.wav", speech["audio"], samplerate=speech["sampling_rate"])
```

# suno/bark

Модель bark знает много языков. Выбрать голос для нее можно тут: https://suno-ai.notion.site/8b8e8749ed514b0cbf3f699013548683?v=bc67cff786b04b50b3ceb756fd05f68c

```python
from transformers import BarkModel
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = BarkModel.from_pretrained("suno/bark-small", torch_dtype=torch.float16).to(device)
```

```python
# Эта фича способствует более быстрой генерации.
model.enable_cpu_offload()
```

```python
# Эта тоже.
model =  model.to_bettertransformer()
```

```python
# Этот код создает модель Bark и генерирует аудио в виде массива чисел,
# который сохраняется в переменную audio_array.
# Затем ее нужно будет конвертировать в звуковой файл и сохрнаить его.
# Работает модель довольно медленно - нужно терпение.
from transformers import AutoProcessor, BarkModel

processor = AutoProcessor.from_pretrained("suno/bark")
model = BarkModel.from_pretrained("suno/bark")

voice_preset = "v2/en_speaker_6"

inputs = processor("Hello, my dog is cute", voice_preset=voice_preset)

audio_array = model.generate(**inputs)
audio_array = audio_array.cpu().numpy().squeeze()
```

```python
# Чтобы записать файл, используем библиотеку scipy.
import scipy
```

```python
# Этот параметр задает частоту семплирования (про которое вы уже знаете,
# потому что читаете мой блог. Но если не читали, тот пост тут: https://t.me/freeAIintro/311).
sampling_rate = model.generation_config.sample_rate
print(sampling_rate)
```

```python
# Создаем и созраняем аудиофайл.
scipy.io.wavfile.write("bark_out.wav", rate=sampling_rate, data=audio_array)
```

```python
# А вот пример генерации для русского языка.
inputs = processor("Баня сгорела. Не парься.", voice_preset='v2/ru_speaker_5')

audio_array = model.generate(**inputs)
audio_array = audio_array.cpu().numpy().squeeze()

scipy.io.wavfile.write("banya.wav", rate=sampling_rate, data=audio_array)
```
